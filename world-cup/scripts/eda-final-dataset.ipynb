```python
wc = df.query('competition == "world-cup"').copy()
wc.shape
```

```python
wc.season.value_counts()
```

```python
# average odd-win by world-cup
cols = ['season','odd_win']
wc[cols].groupby('season').agg(['min', 'max','mean']).reset_index()
```
