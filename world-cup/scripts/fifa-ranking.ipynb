```python
import pandas as pd
```

```python
rf = pd.read_csv('dataset/fifa_ranking-2021-05-27.csv')
# change South Korea when abrv == KOR
for idx, r in rf.query('country_abrv=="KOR"').iterrows():
    rf.loc[idx,'country_full']='South Korea'
    
# change North Korea when abrv == PRK
for idx, r in rf.query('country_abrv=="PRK"').iterrows():
    rf.loc[idx,'country_full']='North Korea'
    
# change Cote d'Ivoire Côte d'Ivoire
for idx, r in rf.query('country_abrv=="CIV"').iterrows():
    rf.loc[idx,'country_full']="Côte d'Ivoire"
  
```

```python
for idx, r in tqdm(odd.iterrows()):
    score = r.correct_score
    score = score.split(':')
    odd.loc[idx,'home_score']=int(score[0])
    odd.loc[idx,'away_score']=int(score[1])
    odd.loc[idx,'diff_score']=abs(int(score[0]) - int(score[1]))
```

```python
odd = pd.read_csv('dataset/game_odd_rank_fifa.csv')
odd['matchday']= pd.to_datetime(odd['matchday'],format='%d.%m.%Y')

# add team points
for idx, r in odd.iterrows():
    if r.home_score>r.away_score:
        odd.loc[idx,'home_points']=3
    elif r.home_score == r.away_score:
        odd.loc[idx,'home_points']=1
    else:
        odd.loc[idx,'home_points']=0
    if r.home_score>r.away_score:
        odd.loc[idx,'away_points']=0
    elif r.home_score == r.away_score:
        odd.loc[idx,'away_points']=1
    else:
        odd.loc[idx,'away_points']=3



unique_home = odd.home_team.unique()
unique_away = odd.away_team.unique()

countries=pd.DataFrame(unique_home.tolist() + unique_away.tolist()).drop_duplicates().loc[:,0].tolist()
```

```python
team_trends = []
#countries = ['France','England','Germany']
for team in countries:
    #team = 'England'
    SWE=odd.query('home_team == @team or away_team == @team').copy()
    SWE.sort_values('matchday', inplace=True)
    for idx, r in SWE.iterrows():
        if r.home_team == team:
            team_trends.append([r.home_team, r.matchday, r.home_points, r.odd_win])
        else:
            team_trends.append([r.away_team, r.matchday, r.away_points, r.odd_win])


    temp_team=pd.DataFrame(team_trends)
    temp_team.columns = ['team','matchday','points','odd_win']
    temp_team['matchday']=pd.to_datetime(temp_team['matchday'], format='%d.%m.%Y')
    temp_team.sort_values('matchday', inplace=True)
    temp_team['rolling_points']=temp_team.points.rolling(6).sum().fillna(-1)
    temp_team['rolling_winning_odd']=temp_team.odd_win.rolling(4).mean().fillna(-1)
    temp_team.sort_values('matchday')



    temp_team['datekey']=temp_team['matchday'].dt.strftime('%Y%m%d')

```
